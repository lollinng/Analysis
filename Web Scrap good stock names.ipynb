```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://stock-financials.valuestocks.in/en/share-market-tips/nse-futures/stocks'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
# class identified by seeing its page source
table = soup.find('table',class_='table table-bordered tb')
```

```python


stock_list = ["FUTURE COFORGE"]

for team in table.find_all('tbody'):
    rows = team.find_all('tr')
    for row in rows:
        stock = row.find('a').text
        print(stock)
        if stock != 'COFORGE':
            stock_list.append(stock)
```

```python
print(stock_list)  
```

```python
stock_list_df = pd.DataFrame(stock_list,columns =['Fruits'])
stock_list_df.head()
```

```python
stock_list_df.to_csv('Future Stock Names.csv')
```
