# Change maker

```python
df['CHANGE'] = ((df['CLOSE'] - df['OPEN'])*100 / df['OPEN']).round(2)
df.sort_values('CHANGE')
df.head()
```

```python
df_filtered = df[df['CLOSE'] >= 100]
df_filtered = df_filtered[df_filtered['TOTALTRADES'] >= 20000]
df_filtered = df_filtered[abs(df_filtered['CHANGE']) >= 3.5]
df_filtered
```

```python
winners = df_filtered[df_filtered['CHANGE']>= 3.5]
print(winners.count().SYMBOL)
winners = winners.sort_values('CHANGE', ascending=False)
winners
```

```python
lossers = df_filtered[df_filtered['CHANGE']<= -3.5]
print(lossers.count().SYMBOL)
lossers = lossers.sort_values(by=['CHANGE'], ascending=True)
lossers
```

```python
losers2 = lossers[['SYMBOL','CHANGE']]
```
