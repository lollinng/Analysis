```python
df = pd.read_csv('cm25AUG2021bhav.csv')
date = df['TIMESTAMP'].iloc[1]
df = df.drop(['HIGH','LOW','LAST','TIMESTAMP','TOTTRDVAL','TOTTRDQTY','ISIN','Unnamed: 13','PREVCLOSE'],axis=1)
df.head(10)
```

# Change maker

```python
df['CHANGE'] = ((df['CLOSE'] - df['OPEN'])*100 / df['OPEN']).round(2)

df.head()
```

```python
df_filtered = df[df['CLOSE'] >= 100]
df_filtered = df_filtered[df_filtered['SERIES'] == 'EQ']
df_filtered = df_filtered[df_filtered['TOTALTRADES'] >= 30000]
df_filtered = df_filtered[abs(df_filtered['CHANGE']) >= 3.5]
df_filtered
```

```python
winners = df_filtered[df_filtered['CHANGE']>= 3.5]
print(winners.count().SYMBOL)
winners = winners.sort_values('TOTALTRADES', ascending=False)
winners.reset_index(inplace=True, drop=True)
winners = winners[['SYMBOL','CHANGE','TOTALTRADES']]
winners
```

```python
lossers = df_filtered[df_filtered['CHANGE']<= -3.5]
print(lossers.count().SYMBOL)
lossers = lossers.sort_values(by=['TOTALTRADES'], ascending=False)
lossers = lossers[['SYMBOL','CHANGE','TOTALTRADES']]
lossers.reset_index(inplace=True, drop=True)
lossers
```
